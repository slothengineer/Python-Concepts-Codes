### 1. Write a program to accept percentage from user and display the grade according to following criteria.



````python
percentage = float(input("Enter the percentage: "))

if percentage > 90:
     grade = 'A'
elif (80 < percentage <= 90):
    grade = 'B'
elif (60 <= percentage <= 80):
    grade = 'C'
else:
    grade = 'F'

print(f"Grade: {grade}")
````

### 2. Write a program to accept cost price of a bike and display road tax to paid according to given criteria



````python
cost_price = float(input("Enter the cost price of the bike: "))

if cost_price > 100000:
    road_tax = cost_price * 0.15
elif (50000 < cost_price <= 100000):
    road_tax = cost_price * 0.10
else:
    road_tax = cost_price * 0.05

print(f"Road tax to pay for the bike: ${road_tax:.2f}")
````

### 3. Accept any city from user and display monuments of that city



````python
# Create a dictionary with city-monument information
city_monuments = {
    'Delhi': ['Red Fort'],
    'Agra': ['Taj Mahal'],
    'Jaipur': ['Jal Mahal'],
}

user_city = input("Enter a city name: ")

if user_city in city_monuments:
    print(f"Monuments in {user_city}:")
    for monument in city_monuments[user_city]:
        print(monument)
else:
    print("Monuments for this city are not available in the database.")
````

### 4. Check how many times a given number can be devided by 3 before  it is less than of equal to 10?

````python
def count_divisions(number):
    count = 0
    while number > 10:
        number //= 3
        count += 1
    return count

given_number = int(input('Enter the number:'))
divisions = count_divisions(given_number)
print(f"The number {given_number} can be divided by 3 {divisions} times before it is less than or equal to 10.")
````

### 5. Why and when to use while loop in python give a detailed description with example ?

A `while` loop in Python is used to repeatedly execute a block of code as long as a specified condition is true. It is a control flow statement that allows you to create loops that continue to run until the condition becomes false. `while` loops are particularly useful when you don't know in advance how many iterations are needed, and you want to keep executing a block of code until a certain condition is met.

Here's a detailed description of when and why to use a `while` loop in Python, along with an example:

**When to Use a `while` Loop:**

1. **Unknown Number of Iterations:** Use a `while` loop when you don't know in advance how many times you need to repeat a block of code. Instead of specifying a fixed number of iterations, you rely on a condition to control when the loop should terminate.

2. **Dynamic Termination:** When the termination condition depends on changing circumstances or user input, a `while` loop is more appropriate. It allows you to keep looping until a specific condition is met.

3. **Continuously Monitoring Conditions:** You can use `while` loops for continuous monitoring or polling tasks, like checking for user input, monitoring sensor data, or waiting for specific events.

**How to Use a `while` Loop with an Example:**

Here's a basic structure of a `while` loop:

```python
while condition:
    # Code to be executed repeatedly
```

Now, let's see an example of a `while` loop:

```python
# Example 1: Counting from 1 to 5 using a while loop
count = 1
while count <= 5:
    print(count)
    count += 1
```

In this example, the `while` loop is used to count from 1 to 5. The condition `count <= 5` checks whether `count` is less than or equal to 5. As long as this condition is true, the loop continues to execute, printing the current value of `count` and incrementing it by 1 in each iteration. The loop stops when the condition becomes false, which happens when `count` reaches 6.

Here's another example that waits for user input and continues to prompt until the user enters 'quit':

```python
# Example 2: Repeatedly prompting the user until 'quit' is entered
user_input = ""
while user_input != "quit":
    user_input = input("Enter 'quit' to exit: ")
```

In this case, the loop continues until the user enters 'quit'. It checks the condition `user_input != "quit"` in each iteration.

Remember that it's important to ensure that the condition in a `while` loop eventually becomes false, or you may create an infinite loop, which can lead to the program running indefinitely.

### 6. Use nested while loop to print three different pattern.

````python
# Pattern 1: Right-angled Triangle
n = 5  
i = 1

while i <= n:
    j = 1
    while j <= i:
        print('*', end=' ')
        j += 1
    print()
    i += 1
````

````python
# Pattern 2: Inverted Right-angled Triangle
n = 5  
i = n

while i >= 1:
    j = 1
    while j <= i:
        print('*', end=' ')
        j += 1
    print()
    i -= 1
````

````python
# Pattern 3: Square
n = 4  
i = 1

while i <= n:
    j = 1
    while j <= n:
        print('*', end=' ')
        j += 1
    print()
    i += 1
````
